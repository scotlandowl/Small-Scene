```python
import os
from collections import Counter
```

```python
root = "./data/Small_Scenes/"
folder_paths = ['bounding_box_test', 'query', 'query_1', 'query_2', 'query_3', 'query_4', 'query_5', 'query_6', 'query_7', 'query_8', 'query_9', 'query_10', 'query_11']
for i, x in enumerate(folder_paths):
    folder_path = root + x
    file_list = os.listdir(folder_path)
    
    cnt = Counter()
    for file in file_list:
        f_id = int(file.split("_")[0])
        if cnt[f_id] != i:
            file_name = folder_path + '/' + file
            os.remove(file_name)
        cnt[f_id] += 1
    
```
